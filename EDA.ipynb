```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

df = pd.read_csv("/content/drive/MyDrive/College/Stock_Market_Predictor/GOOG_train.csv")
```

```python
plt.rcParams["figure.figsize"] = [15,10]
plt.rcParams["figure.autolayout"] = True
plt.plot(df.iloc[:,0], df.iloc[:,1],'b', label = 'Opening Prices')
plt.plot(df.iloc[:,0], df.iloc[:,2],'g', label = 'Highest Prices')
plt.plot(df.iloc[:,0], df.iloc[:,3],'r', label = 'Lowest Prices')
plt.plot(df.iloc[:,0], df.iloc[:,4],'y', label = 'Closing Prices')
plt.xlabel('Time')
plt.ylabel('Prices')
plt.title('Google Stock Prices', fontsize = 20)
plt.legend()
plt.show()
```

We can see that google stock prices increases almost continously with a drop in between.

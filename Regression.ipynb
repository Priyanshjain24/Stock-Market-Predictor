```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn

df = pd.read_csv("/content/drive/MyDrive/College/Stock_Market_Predictor/GOOG_train.csv")
```

```python
X = df[[ 'Open' , 'High' , 'Low' , 'Volume' ]]
y = df[ 'Close' ]

from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
training_set_scaled = SS.fit_transform(y_train.array.reshape(-1,1), X_train)

from sklearn.linear_model import LinearRegression
linreg = LinearRegression().fit(X_train, training_set_scaled)
predictions = linreg.predict(X_train)
```

```python
from matplotlib import pyplot
pyplot.scatter(y_train, training_set_scaled)
pyplot.plot(np.sort(y_train, axis=0),predictions)
pyplot.show()
```

```python
sklearn.metrics.r2_score(y_train,predictions)
```
